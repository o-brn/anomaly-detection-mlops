```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import plotly.graph_objs as go
```

```python
df = pd.read_csv(r"..\data\raw\processed.csv")
df.dtypes
```

# Modelo

```python
# Instanciar o modelo
model = IsolationForest(contamination="auto", random_state=42)

# Treinar o modelo
model.fit(df[["value"]])

# Fazer predições (1 para inliers, -1 para outliers)
df["anomaly"] = model.predict(df[["value"]])
```

```python
# Série principal
trace_main = go.Scatter(x=df.index, y=df["value"], mode="lines", name="Utilização de CPU")

# Pontos de anomalia
anomalies = df[df["anomaly"] == -1]
trace_anomalies = go.Scatter(
    x=anomalies.index,
    y=anomalies["value"],
    mode="markers",
    name="Anomalia Detectada",
    marker=dict(color="red", size=8),
)

fig = go.Figure(data=[trace_main, trace_anomalies])

fig.update_layout(
    title="Detecção de Anomalias com Isolation Forest",
    xaxis_title="Timestamp",
    yaxis_title="Utilização (%)",
    template="plotly_white",
    width=900,
    height=400,
    legend=dict(x=0, y=1),
)

fig.show()

print(f"Número de anomalias detectadas: {len(anomalies)}")
```
